### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns

# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
plt.figure(figsize=(12,8))
ax = sns.histplot(data=df, x='age', palette='cividis', bins=12)
plt.xlabel('Age', fontsize=12)
plt.ylabel('# of Managers', fontsize=12)
plt.title('Ages of Managers in Survey', fontsize=16)
```

##### Insights:

The managers who took the survey were aged from about 20-65, with the highest numbers being from around 50-65, although the numbers were similar throughout starting at around age 30.

### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(12,6))
ax = sns.histplot(data=df, x='industry', palette='cividis', bins=10)
plt.xlabel('Industry', fontsize=12)
plt.ylabel('# of Managers', fontsize=12)
plt.title('Industry of Managers in Survey', fontsize=16)
```

##### Insights:

The amount of managers in each industry is almost equal, with the exception being healthcare which is a little lower. The four industries are business, finance, technology, and healthcare.

### Plot for 'salary' column

```python
# create plot for salary column
plt.figure(figsize=(12,6))
ax = sns.distplot(df['salary'], kde=False)
plt.xlabel('Salary', fontsize=12)
plt.ylabel('# of Managers', fontsize=12)
plt.title('Distribution of Salary Among Managers in Survey', fontsize=16)
```

##### Insights:

The salary for managers in the survey ranges from about 20,000 to over 160,000, with the majority lying in the 80,000 to 140,000 range. The most common number is around 90,000-100,000.

### Plot for 'industry_experience' column

```python
# create plot for industry experience column
plt.figure(figsize=(12,6))
ax = sns.histplot(data=df, x='industry_experience', bins=20)
plt.xlabel('Industry Experience', fontsize=12)
plt.ylabel('# of Managers', fontsize=12)
plt.title('Industry Experience for Managers Taking Survey in Years', fontsize=16)
```

##### Insights:

The industry experience for managers range from 0 to 35 years, with 10-25 being the most popular range.

### Plot for 'education' column

```python
# create plot for education levels
plt.figure(figsize=(10,10))
data = df['education'].value_counts()
labels = ["Bachelor's Degree", "Master's Degree", "Some College", "High School"]
colors = sns.color_palette('Accent')
plt.title('Manager\'s Education Levels', fontsize=16)
ax = plt.pie(data, explode=None, colors = colors, labels = labels, autopct = '%.0f%%')
```

##### Insights
The most managers had a Bachelor's Degree, while a third of them had a Master's Degree.

### Plot for 'gender' column

```python
# create plot for gender column
plt.figure(figsize=(10,10))
data = df['gender'].value_counts()
labels = ['Male', 'Female']
colors = sns.color_palette('bright')
plt.title('Gender of Managers')
ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%')
```

##### Insights:

The gender of managers is split in half, with half being male, and the other being female.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df = pd.read_csv('salary_numeric.csv')
df.head()
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = [...]

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df['salary'], df['education'])
```

There is not a strong correlation between salary and education. The correlation is only 0.48 and the p value is 0.0. I would say they are correlated but there is not a strong correlation between them.

#### Are these results statistically significant? How do you know?

No, because there is not a strong correlation between them.

### What is the correlation between salary and age?

#### Your Hypothesis [...]

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df['salary'], df['age'])
```

There is a relatively strong correlation between salary and age, as the correlation is at 0.66. I still would not say it is very strong, but it is relatively strong.

#### Are these results statistically significant? How do you know? 

Yes, because it is a stronger correlation between them.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = [...]

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(df['salary'], df['industry_experience'])
```

There is a decently strong correlation between salary and industry experience.

#### Are these results statistically significant? How do you know?

I would say no, because the number is still not close enough to 1 to be statistically significant.

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(2)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
c = pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(12,6))
ax = sns.heatmap(c, annot=True, cmap='Accent', fmt='g', xticklabels=['High School', 'Some College', 'Bachelors Degree', 'Masters Degree'])
plt.title('Salary for Managers Across Industries by Education Level', fontsize=16)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Industry', fontsize=12)
```

##### Insights:

The strong correlations are with masters degree and technology as the industry, as well as masters degree and finance. There were strong correlations for healthcare and business with masters degree as well. Bachelors and technology was also very strong. There were weak correlations with each industry for high school education level.
